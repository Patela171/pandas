```python
import pandas as pd
```

```python
# Load Training CSV
projects = pd.read_csv("./Tweets.csv", index_col=False)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# Vectorize Titles: Make them into arrays that can be used for training
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(projects["text"])
# View shape of the Vectorized Titles
X_train_counts.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_train_counts, projects["airline_sentiment"], test_size=0.3, shuffle=True) # Splitting into a train set and test set
```

```python
from sklearn.naive_bayes import MultinomialNB
# Train model with training data
clf = MultinomialNB().fit(X_train, y_train)
```

```python
from sklearn import metrics
# View performance on training data
predicted = clf.predict(X_test)
print(metrics.classification_report(y_test, predicted, zero_division=False))
# It doesnt look too bad! Around 70% seems alright for the amount of training data that we had available
```
