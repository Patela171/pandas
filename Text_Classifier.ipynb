```python
import pandas as pd
```

```python
# Load Training CSV
projects = pd.read_csv("./project_names.csv", index_col=False)
```

```python
projects.columns = ["Area","Title"] # Override Column Headers
```

```python
projects # Look at the Projects
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# Vectorize Titles: Make them into arrays that can be used for training
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(projects['Title'])
# View shape of the Vectorized Titles
X_train_counts.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
# Train model with training data
clf = MultinomialNB().fit(X_train_counts, projects['Area'])
```

```python
from sklearn import metrics
# View performance on training data
predicted = clf.predict(X_train_counts)
print(metrics.classification_report(projects['Area'], predicted, zero_division=False))
# It doesnt look too bad! Around 70% seems alright for the amount of training data that we had available
```

```python
# Classify our titles!
cysf = pd.read_csv("cysf_awards.csv")
cysf
```

```python
# Tansform titles into the same format as the training data
cysf_title_counts = count_vect.transform(cysf['TITLE'])

# Predict new areas
predicted_cysf = clf.predict(cysf_title_counts)

# Add predicted areas to the table
cysf['Predicted Area'] = predicted_cysf
```
