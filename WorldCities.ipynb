```python
import pandas as pd
```

```python
df = pd.read_csv("worldcitiespop.csv")
```

```python
print(df.columns.tolist())
Canada = df[df["country"]=='ca']
```

```python
Canada[Canada["region"]==8].sort_values("population", ascending=False)
```

#Find the top 5 biggest cities in Ontario.

**Toronto, Ottawa, Hamilton, Kitchener, London**

```python
CanadaGroup = Canada.sort_values("Population", ascending = False).groupby("Region")
```

```python
CanadaGroup.head(1).sort_values("Region")
```
