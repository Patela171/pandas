```python
# Import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Create an URL object
url = 'https://www.cysf.org/award-winners/'
# Create object page
page = requests.get(url)
```

```python
# parser-lxml = Change html to Python friendly format
# Obtain page's information
soup = BeautifulSoup(page.text, 'lxml')
soup
```

```python
# Obtain information from tag <table>
table1 = soup.find('table', id='tablepress-6')
table1
```

```python
# Obtain every title of columns with tag <th>
headers = []
for i in table1.find_all('th'):
 title = i.text
 headers.append(title)
```

```python
# Create a dataframe
mydata = pd.DataFrame(columns = headers)
mydata
```

```python
# Create a for loop to fill mydata
for j in table1.find_all('tr')[1:]:
 row_data = j.find_all('td')
 row = [i.text for i in row_data]
 length = len(mydata)
 mydata.loc[length] = row
```

```python
mydata
```

```python
# Export to csv
mydata.to_csv('CYSF_Award_Winners.csv', index=False)

# Try to read csv
mydata2 = pd.read_csv('CYSF_Award_Winners.csv')
```

```python
# Cleaning up the dataset

mydata2 = mydata2.fillna(" ")

# Set Pandas to display the entire dataset
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Create a column to determine which field gets most prizes
new_data = [1] * 122
# Add the new column to the DataFrame
mydata2['Count'] = new_data
```

```python
mydata2.groupby("TITLE").sum().sort_values("Count", ascending=False)
```

```python
mydata2.groupby("AWARDNAME").sum().sort_values("Count", ascending=False)
```
